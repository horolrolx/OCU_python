```python
score = pd.read_excel("C:/OCU교육/data/성적처리.xlsx", sheet_name = 'Sheet1')
score.head()
```

```python
score = score.dropna()
score["합계"] = score.sum(1)
score.head()
```

```python
score1 = score.iloc[:, [0, 2, 3, 4, 5, 6]]
ban = pd.DataFrame(index = ['국어', '영어', '수학', '사회', '과학'])
ban["1반"] = score1.loc[score1["반"] == "1반"].sum()
ban["2반"] = score1.loc[score1["반"] == "2반"].sum()
ban["3반"] = score1.loc[score1["반"] == "3반"].sum()
ban["4반"] = score1.loc[score1["반"] == "4반"].sum()
ban["5반"] = score1.loc[score1["반"] == "5반"].sum()
ban
```

```python
score["평균"] = score.loc[:, "국어":"과학"].mean(1)
score["평균1"] = score.iloc[:, 2:7].mean(1)
score
```

```python
score1 = score.iloc[:, [0, 2, 3, 4, 5, 6]]
aver = pd.DataFrame(index = ["국어","영어","수학","사회","과학"])
aver["1반"] = score1.loc[score1["반"] == "1반"].mean()
aver["2반"] = score1.loc[score1["반"] == "2반"].mean()
aver["3반"] = score1.loc[score1["반"] == "3반"].mean()
aver["4반"] = score1.loc[score1["반"] == "4반"].mean()
aver["5반"] = score1.loc[score1["반"] == "5반"].mean()
aver
```

```python
import numpy as np

score["평균_반올림"] = score["평균"].round(1)
score["평균_올림"] = score["평균"].apply(np.ceil)
score["평균_내림"] = score["평균"].apply(np.floor)
score["평균_버림"] = score["평균"].apply(np.trunc)
score
```

```python
score["순위"] = score["평균"].rank(ascending = False)
score["순위_min"] = score["평균"].rank(method = 'min', ascending = False)
score["순위_max"] = score["평균"].rank(method = 'max', ascending = False)
score["순위_first"] = score["평균"].rank(method = 'first', ascending = False)
score["순위_dense"] = score["평균"].rank(method = 'dense', ascending = False)
score
```

```python
score_row = score.iloc[:, [0, 2, 3, 4, 5, 6]]
score["MIN"] = score_row.min(1)
score["MAX"] = score_row.max(1)
score.head()
```

```python
score_row = score.iloc[:, [0, 2, 3, 4, 5, 6]]
score["MEDIAN"] = score_row.median(1)
score.head()
```
