```python
left = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                    'key2': ['K0', 'K1', 'K0', 'K1'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                     'key2': ['K0', 'K0', 'K0', 'K0'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'left')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'right')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'outer')
result
```

```python
import pandas as pd

df = pd.read_excel('C:/OCU교육/data/회원정보.xlsx', sheet_name = 'Sheet1')
df = df.dropna()
df['성별구분'] = df['주민등록번호'].str[7]
mapping = {'1':'남성','2':'여성','3':'기타','4':'기타'}
df['성별'] = df['성별구분'].map(mapping)
df
```

```python
for i in range(len(df['주민등록번호'])):
    if df['주민등록번호'].str[7].loc[i] == '1':
        df['성별1'].loc[i] = '남성'
    elif df['주민등록번호'].str[7].loc[i] == '2':
        df['성별1'].loc[i] = '여성'
    else:
        df['성별1'].loc[i] = '기타'
df
```

```python
customer = df
pay = pd.read_excel('C:/OCU교육/data/회원정보_연봉.xlsx', sheet_name = '연봉')
pay = pay.dropna()
pay
```

```python
pay = pay[['성명','연봉']]
customer['성명'] = customer['성'] + customer['이름']
customer_pay = pd.merge(customer, pay, on ='성명', how = 'outer')
customer_pay = customer_pay.drop(['성별'], axis = 1)
customer_pay.head()
```
