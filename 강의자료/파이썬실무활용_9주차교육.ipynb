```python
import pandas as pd
import numpy as np

df = pd.read_excel("../data/회원정보.xlsx", sheet_name = 'Sheet1')
df
```

```python
df = df.dropna()
```

```python
df['성명'] = df['성'] + df['이름']
df['성명1'] = df[['성','이름']].sum(1)
df
```

```python
df['주민번호앞6자리'] = df['주민등록번호'].str[0:6]
df['성별'] = df['주민등록번호'].str[7]
df['전화번호뒤4자리'] = df['전화번호'].str[9:14]
df
```

```python
df[['성','이름','주민등록번호','전화번호','주민번호앞6자리','성별','전화번호뒤4자리']]
```

```python
df['대글자'] = df['영문명'].str.upper()
df['첫글자'] = df['영문명'].str.capitalize()
df
```

```python
df['phone'] = df['전화번호'].str.replace("-","")
df['phone1'] = df['전화번호'].str.replace("-","",1)
df['phone2'] = df['전화번호'].str.replace("-","",2)
df
```

```python
df['FIND'] = df['주민등록번호'].str.find("-")
df['성별구분'] = df['주민등록번호'].str[df['주민등록번호'].str.find("-")[0] + 1]
df
```

```python
df['이름수'] = df['이름'].str.len()
df
```

```python
df['주소(공백제거)'] = df['주소'].str.strip()
df['주소_len'] = df['주소'].str.len()
df['주소_len2'] = df['주소(공백제거)'].str.len()
df
```
