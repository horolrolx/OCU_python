```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']},
                  index = [0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B': ['B4', 'B5', 'B6', 'B7'],
                   'C': ['C4', 'C5', 'C6', 'C7'],
                   'D': ['D4', 'D5', 'D6', 'D7']},
                  index = [4, 5, 6, 7])
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                   'B': ['B8', 'B9', 'B10', 'B11'],
                   'C': ['C8', 'C9', 'C10', 'C11'],
                   'D': ['D8', 'D9', 'D10', 'D11']},
                  index = [8, 9, 10, 11])
```

```python
result = pd.concat([df1, df2, df3])
result
```

```python
result = pd.concat([df1, df2, df3], axis = 1)
result
```

```python
result = pd.concat([df1, df2, df3], axis =1, ignore_index = True)
result
```

```python
result.rename(columns = {0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',9:'J',10:'K',11:'L'}, inplace = True)
result
```

```python
left = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                    'key2': ['K0', 'K1', 'K0', 'K1'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                     'key2': ['K0', 'K0', 'K0', 'K0'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})
result = pd.merge(left, right, on = ['key1', 'key2'])
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'left')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'right')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'outer')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'left', indicator = True)
result
```
